```python
import pandas as pd
final_df = pd.read_csv(r"/content/combined_cleaned_df_final.csv")
```

```python
final_df = final_df.drop('Unnamed: 0', axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

gender_encoder = LabelEncoder()

final_df['gender'] = gender_encoder.fit_transform(final_df['gender'])
```

```python
df_icd9 = final_df[final_df['icd_version'] == 9]
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
icd_encoded = encoder.fit_transform(df_icd9[['icd_code']])
icd_encoded_df = pd.DataFrame(
    icd_encoded,
    columns=encoder.get_feature_names_out(['icd_code'])
)

icd_encoded_df
```

```python
y = icd_encoded_df
y
```

```python
X = df_icd9.drop(['hadm_id', 'icd_code', 'subject_id'], axis=1)
X
```

```python
y_labels = y.idxmax(axis=1)
X.columns = X.columns.astype(str)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled
```

```python
import numpy as np
y_class = np.argmax(y.values, axis=1)
y_class
```

```python
X_scaled.shape
```

```python
import xgboost as xgb

model_xgb = xgb.XGBClassifier(
    objective='multi:softmax',
    num_class=len(y.columns),
    eval_metric='mlogloss',
    n_estimators=150,
    random_state=42
)

model_xgb.fit(X_scaled, y_class)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve

# Generate predictions using the trained XGBoost model
y_pred = model_xgb.predict(X_scaled)

# Compute Accuracy
accuracy = accuracy_score(y_class, y_pred)

# Compute Precision, Recall, and F1-score
precision = precision_score(y_class, y_pred, average='weighted')  # Use 'macro' or 'micro' if needed
recall = recall_score(y_class, y_pred, average='weighted')
f1 = f1_score(y_class, y_pred, average='weighted')

# Compute ROC AUC Score (only applicable for binary classification or using probabilities)
if len(set(y_class)) == 2:  # Ensure it's a binary classification problem
    y_pred_prob = model_xgb.predict_proba(X_scaled)[:, 1]  # Get probabilities for the positive class
    roc_auc = roc_auc_score(y_class, y_pred_prob)

    # Compute ROC Curve
    fpr, tpr, _ = roc_curve(y_class, y_pred_prob)
else:
    roc_auc = None  # AUC is not directly applicable for multi-class

# Print Metrics
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")

# Print ROC AUC Score for binary classification
if roc_auc is not None:
    print(f"ROC AUC Score: {roc_auc:.2f}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# ---------------------------------------------------------------------------------
# 1. Convert NumPy arrays to PyTorch tensors and create a DataLoader
# ---------------------------------------------------------------------------------

# Convert feature matrix (X_scaled) and labels (y_class) to PyTorch tensors
X_tensor = torch.from_numpy(X_scaled).float()
y_tensor = torch.from_numpy(y_class).long()

# Create a dataset using TensorDataset
dataset = TensorDataset(X_tensor, y_tensor)

# DataLoader for batch processing during training (batch size = 32, shuffling enabled)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

# ---------------------------------------------------------------------------------
# 2. Define the Neural Network architecture
# ---------------------------------------------------------------------------------

class SimpleNet(nn.Module):
    def __init__(self, input_size=25, hidden1=256, hidden2=512, hidden3=256, hidden4=128, num_classes=1203):
        super(SimpleNet, self).__init__()

        # Fully connected layers with batch normalization and ReLU activations
        self.fc1 = nn.Linear(input_size, hidden1)
        self.bn1 = nn.BatchNorm1d(hidden1)
        self.relu1 = nn.ReLU()

        self.fc2 = nn.Linear(hidden1, hidden2)
        self.bn2 = nn.BatchNorm1d(hidden2)
        self.relu2 = nn.ReLU()

        self.fc3 = nn.Linear(hidden2, hidden3)
        self.bn3 = nn.BatchNorm1d(hidden3)
        self.relu3 = nn.ReLU()

        self.fc4 = nn.Linear(hidden3, hidden4)
        self.bn4 = nn.BatchNorm1d(hidden4)
        self.relu4 = nn.ReLU()

        # Final output layer (no activation because CrossEntropyLoss expects raw logits)
        self.fc5 = nn.Linear(hidden4, num_classes)

    def forward(self, x):
        """Defines the forward pass of the network."""
        x = self.relu1(self.bn1(self.fc1(x)))
        x = self.relu2(self.bn2(self.fc2(x)))
        x = self.relu3(self.bn3(self.fc3(x)))
        x = self.relu4(self.bn4(self.fc4(x)))
        x = self.fc5(x)  # Output logits
        return x

# ---------------------------------------------------------------------------------
# 3. Model Initialization and Training Setup
# ---------------------------------------------------------------------------------

# Check if GPU is available and use it, otherwise fallback to CPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Initialize the model and move it to the chosen device
model = SimpleNet(input_size=25, num_classes=1203).to(device)

# Define the loss function (CrossEntropyLoss is used for multi-class classification)
criterion = nn.CrossEntropyLoss()

# Define the optimizer (Adam optimizer with weight decay for regularization)
optimizer = optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)

# Learning rate scheduler to reduce LR every 10 epochs by a factor of 0.9
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.9)

# ---------------------------------------------------------------------------------
# 4. Training Loop
# ---------------------------------------------------------------------------------

num_epochs = 100  # Total number of training epochs

for epoch in range(num_epochs):
    model.train()  # Set the model to training mode
    running_loss = 0.0  # Variable to track loss per epoch

    for batch_X, batch_y in dataloader:
        # Move batch data to the appropriate device (CPU/GPU)
        batch_X, batch_y = batch_X.to(device), batch_y.to(device)

        optimizer.zero_grad()  # Reset gradients before backpropagation
        outputs = model(batch_X)  # Forward pass
        loss = criterion(outputs, batch_y)  # Compute loss

        loss.backward()  # Backpropagation
        optimizer.step()  # Update weights

        running_loss += loss.item()  # Accumulate batch loss

    avg_loss = running_loss / len(dataloader)  # Compute average loss for the epoch
    scheduler.step()  # Adjust learning rate according to schedule

    # Print loss after every epoch
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}")

# ---------------------------------------------------------------------------------
# 5. Save the Trained Model
# ---------------------------------------------------------------------------------

# Save the trained model parameters to a file
torch.save(model.state_dict(), "model.pth")
print("Training complete. Model saved as 'model.pth'.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import torch

# ---------------------------------------------------------------------------------
# 6. Model Evaluation
# ---------------------------------------------------------------------------------

# Load the trained model for evaluation
model.eval()  # Set the model to evaluation mode

# Move tensors to the appropriate device
X_tensor, y_tensor = X_tensor.to(device), y_tensor.to(device)

# Perform inference (get model predictions)
with torch.no_grad():
    logits = model(X_tensor)  # Get raw logits
    y_pred = torch.argmax(logits, dim=1).cpu().numpy()  # Convert logits to predicted class indices
    y_true = y_tensor.cpu().numpy()  # Convert ground-truth tensor to numpy array

# Compute evaluation metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='weighted')  # Use 'weighted' for class imbalance
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')

# Print the results
print(f"Evaluation Results:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import torch
import numpy as np

# ---------------------------------------------------------------------------------
# 6. Model Evaluation
# ---------------------------------------------------------------------------------

# Set the model to evaluation mode
model.eval()

# Move tensors to the appropriate device
X_tensor, y_tensor = X_tensor.to(device), y_tensor.to(device)

# Perform inference (get model predictions)
with torch.no_grad():
    logits = model(X_tensor)  # Get raw logits
    y_pred = torch.argmax(logits, dim=1).cpu().numpy()  # Convert logits to predicted class indices
    y_true = y_tensor.cpu().numpy()  # Convert ground-truth tensor to numpy array
    y_prob = torch.softmax(logits, dim=1).cpu().numpy()  # Convert logits to probabilities

# Compute evaluation metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='weighted')  # 'weighted' accounts for class imbalance
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')

# Compute ROC AUC score using the one-vs-rest (OvR) strategy
roc_auc = roc_auc_score(y_true, y_prob, multi_class='ovr', average='weighted')

# Print the results
print(f"Evaluation Results:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"ROC AUC Score: {roc_auc:.4f}")
```
